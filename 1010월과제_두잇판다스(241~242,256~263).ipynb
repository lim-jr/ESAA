# 12-1. datetime 오브젝트


* datetime 라이브러리
> a. date 오브젝트 : 날짜 처리  
> b. time 오브젝트 : 시간 처리  
> c. datetime 오브젝트 : 날짜와 시간을 모두 처리

```python
# datetime 라이브러리 불러오기

from datetime import datetime
```

```python
# now, today 메소드 : 날짜와 시간 출력

now1 = datetime.now()
print(now1)

now2 = datetime.today()
print(now2)
```

```python
# 시간을 직접 입력하여 인자로 전달 

t2 = datetime(1920, 1, 1)
t3 = datetime(1999, 6, 1, 3, 34, 59)

print(t2)
print(t3)
```

```python
# 시간 계산을 할 수 있음

diff1 = t2-t3
print(diff1)
print(type(diff1))

diff2= t3-t2
print(diff2)
print(type(diff2))
```

## 2) datetime 오브젝트를 인덱스를 설정해 데이터 추출하기 

```python
# Date열을 tesla의 인덱스로 지정
tesla.index = tesla['Date']
print(tesla.index)
```

```python
tesla.Date.dt.year.value_counts()
# 책엔 2010~2015년이고, 여기엔 2018, 2017년 데이터만 있음
```

```python
# datetime오브젝트를 인덱스로 지정했을 때, 데이터추출하는 방법

# 1. 2018년 데이터 뽑고, 4개행x5개열만 뽑아라
print(tesla['2018'].iloc[:4, :5])

# 2. 2018년 1월 데이터 추출
print(tesla['2018-01'].iloc[:,:5])
```

## 3) 시간 간격을 인덱스로 지정해 데이터 추출하기

```python
# Date열 - Date열의 최솟값 : 데이터 수집한 이후에 시간이 얼마나 흘렀는지 알 수 있음.
# '시간간격'을 나타내는 ref_date열을 추가한다. 

tesla['ref_date'] = tesla['Date']-tesla['Date'].min()
print(tesla.head())
```

```python
# ref_date열을 인덱스로 지정해서, 
# 시간간격을 이용해 데이터를 추출할 수 있음

tesla.index = tesla['ref_date']
tesla.iloc[:5, :5]
## 큰수부터 작은수로 내림차순으로 정렬됨
```

```python
# 데이터수집이후 최초 5일의 데이터를 추출
tesla['5 days': ].iloc[:5, :5]
## 5days: = 5,4,3,2,1,0 
## 내림차순으로 감, 위에서 밑으로 감.
```

```python
tesla[:'5 days'].iloc[:5, :5]
## :5days = 167,166...7,6
## 위에서 밑으로 내려옴
```
