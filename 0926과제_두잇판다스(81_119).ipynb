### 1. 데이터 시각화 개괄

#### (1) 앤스콤 4분할 그래프
* 앤스콤 4분할 그래프를 구성하는 데이터 집합은 4개의 그룹으로 구성
* 4개의 그룹 모두 x,y열을 가지고 있음

```python
## 1. 앤스콤 데이터 집합 불러오기

import seaborn as sns
# 앤스콤 데이터 집합은 seaborn 라이브러리에 포함되어 있음

anscombe = sns.load_dataset("anscombe")
# sns 라이브러리의 load_dataset 메서드에 문자열 anscombe를 전달하면
# 앤스콤 데이터 집합을 불러올 수 있음.

print(anscombe)
# 'dataset'라는 열은 데이터그룹을 구분
print(type(anscombe))
# 데이터프레임 타입
```

```python
## 2. matplotlib 라이브러리로 그래프 그리기
## plot메소드

%matplotlib notebook 
#주피터 노트북에서 바로 실행
%matplotlib inline  
#노트북 브라우저 내부에서 그래프 그려라
import matplotlib.pyplot as plt
# matplotlib 라이브러리 불러오기

dataset1 = anscombe[anscombe['dataset']=='I']
# dataset열에서 데이터값이 I 인것만 추출

plt.plot(dataset1['x'], dataset1['y'])
# x축 y축을 전달
# 선으로 그래프 그려라 (기본값)

plt.plot(dataset1['x'], dataset1['y'], 'o')
# 점으로 그래프 그려라
```

#### (2) 앤스콤 데이터 집합 

1. 전체 그래프가 위치할 기본 틀을 만든다.
2. 그래프를 그려넣을 격자를 만든다.
3. 격자에 그래프를 하나씩 추가한다. 순서는 왼->오
4. 만약 격자의 1행이 꽉차면 2행에 그래프를 그려 넣는다. 

```python
dataset1 = anscombe[anscombe['dataset']=='I']
dataset2 = anscombe[anscombe['dataset']=='II']
dataset3 = anscombe[anscombe['dataset']=='III']
dataset4 = anscombe[anscombe['dataset']=='IV']
# 그룹화된 데이터 추출

fig = plt.figure()
# 1. 기본틀
# figure 메소드

axes1= fig.add_subplot(2,2,1)
axes2= fig.add_subplot(2,2,2)
axes3= fig.add_subplot(2,2,3)
axes4= fig.add_subplot(2,2,4)
# 2. 격자 
# add_subplot메소드 (행, 열, 위치)

axes1.plot(dataset1['x'], dataset1['y'], 'o')
axes2.plot(dataset2['x'], dataset2['y'], 'o')
axes3.plot(dataset3['x'], dataset3['y'], 'o')
axes4.plot(dataset4['x'], dataset4['y'], 'o')
fig
# 3. 그래프 하나씩 추가
# plot메소드

axes1.set_title("dataset1")
axes2.set_title("dataset2")
axes3.set_title("dataset3")
axes4.set_title("dataset4")
fig
# 그래프 격자에 제목 추가
# set_title메소드

fig.suptitle("Anscombe Data")
# 기본 틀에 제목 추가

fig.tight_layout()
# 그래프의 이름과 숫자가 겹쳐보일때 레이아웃 조절
# tight_layout 메소드
```

#### (2) 다변량 그래프 그리기

```python
# 1. 산점도 그래프에서 성별을 색상으로 구분하기

# 1-1. 성별이라는 문자열을 정수로 치환하는 함수
def recode_sex(sex) :
  if sex =='Female':
    return 0
  else:
    return 1

# 1-2. tips 에 새로운 열을 추가
tips['sex_color'] = tips['sex'].apply(recode_sex)
tips.head()

# 2. 산점도 그래프에서 테이블당인원수(size)를 점의 크기로 표현하기
fig = plt.figure()
axes1= fig.add_subplot(1,1,1)

axes1.scatter(
    x=tips['total_bill'],
    y=tips['tip'],
    s=tips['size']*10,   # s: 점의 크기
    c=tips['sex_color'], # c: 점의 색상
    alpha=0.5)           # alpha: 점의 투명도
```

#### (2) 이변량 그래프

##### (a) 산점도 그래프

```python
ax = plt.subplots()
ax = sns.regplot(x='total_bill', y='tip', data=tips)
# regplot메소드로 산점도 그래프 그리기
# 이때 산점도+회귀선 같이 그려짐

ax = sns.regplot(x='total_bill', y='tip', data=tips, fit_reg=False)
## fit_reg옵션으로 회귀선 제거할수 있음

ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
ax1 = sns.lmplot(x='total_bill', y='tip', data=tips, fit_reg=True)
# lmplot 메소드도 가능
```

```python
# joint 메소드 : 산점도 + 히스토그램 한꺼번에 그림

joint = sns.jointplot(x='total_bill', y='tip', data=tips)
joint.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize=10, y=1.05) 
## y옵션 : 제목과 그래프 사이 거리
```

##### (f) 관계 그래프

```python
fig = sns.pairplot(tips)
# pairplot 메소드
# 대각선 : 히스토그램/막대그래프 (변수 1개에 대한 도수)
# 대각선 위/아래 : 산점도
```

```python
pair_grid = sns.PairGrid(tips)
# 관계그래프 그리기 위한 기본틀 만들기
# PairGrid 메소드
pair_grid = pair_grid.map_diag(sns.distplot, rug=True)
# 대각선 (map_diag 메소드)
# : 히스토그램, 확률밀도함수, 러그함수 (displot메소드)
# 
pair_grid = pair_grid.map_upper(sns.regplot)
# 대각선 위 (map_upper 메소드)
# : 산점도, 회귀선 (regplot메소드)
pair_grid = pair_grid.map_lower(sns.kdeplot)
# 대각선 아래 (map_lower 메소드)
# : 이차원밀집도 (kdeplot 메소드)
plt.show()
```
