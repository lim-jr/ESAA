### (1) 레이블 인코딩

### (2) 피처 결측치 처리

#### (2)-1. ints_id, sgg

```python
# inst_id : 인덱스 중복이므로 열 버리기
# sgg : sido와 정보가 겹치므로 열 버리기
train.drop(['inst_id','sgg'],axis=1,inplace=True)
test.drop(['inst_id','sgg'],axis=1,inplace=True)
```

#### (2)-2. 문자열 피처

```python
## 문자열 피처중 instkind, ownerChange 에서 결측치 존재

# instkind(문자열) : 병원종류 결측치 있는 행 다지우기
train.dropna(subset=['instkind'],axis=0,inplace=True)
test.dropna(subset=['instkind'],axis=0,inplace=True)

# ownerChange(문자열) : 대표자변동 결측치있는 행을 최빈값으로 대체
train['ownerChange'].mode()[0]
train.loc[train['ownerChange'].isna(),'ownerChange'] = train['ownerChange'].mode()[0]
train['ownerChange'].isna().sum() #확인
test['ownerChange'].mode()[0]
test.loc[test['ownerChange'].isna(),'ownerChange'] = test['ownerChange'].mode()[0]
test['ownerChange'].isna().sum()
```

#### (2)-3. 숫자형 피처중 금융변수

```python
## 숫자형 피처중 금융변수에서 결측치 존재

# open&같은행번호에서 공통적으로 8개의 결측치가 나타난다. 
print(train[train['revenue1'].isnull()]['OC'])
print(train[train['shortLoan2'].isnull()]['OC'])

# open&같은행번호에서 공통적으로 2개의 결측치가 나타난다.
print(test[test['revenue1'].isnull()])
print(test[test['shortLoan2'].isnull()])

# 따라서 결측치가 존재하는 전체행 지우기
train.dropna(subset=['revenue1'],axis=0,inplace=True)
test.dropna(subset=['revenue1'],axis=0,inplace=True)
```

#### (2)-5. employee

* 매출액 증가율 = 당기매출액 / 전기매출액 *100 -100
* 총자산 증가율 = 당기총자산 / 전기총자산 *100 -100
* 자기자본순이익률(ROE) = 당기순이익/자기자본*100


```python
# 매출액 증가율 = 당기매출액 / 전기매출액 *100 -100
#                  revenue1 / revenue2
# 총자산 증가율 = 당기총자산 / 전기총자산 *100 -100
#            (netAsset1+debt1)/(netAsset2+debt2)
# 순자산총계 = 자산총계-부채총계
# 자산 = 부채+자본
# 자산총계 = 순자산총계 + 부채총계

# 자기자본순이익률(ROE) = 당기순이익/총자본*100
#                          profit1 / (netAsset1) 
# 자기자본 = 자본금+이익잉여금
```

```python
train['매출액증가율']=(train['revenue1']/train['revenue2'])*100-100
train['총자산증가율']=(train['netAsset1']+train['debt1'])/(train['netAsset2']+train['debt2'])*100 -100
train['ROE']=(train['profit1']/train['netAsset1'])*100
```

```python
print(train['매출액증가율'])
print(train['총자산증가율'])
print(train['ROE'])
```
